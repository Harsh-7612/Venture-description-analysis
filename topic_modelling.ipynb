```python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer("all-MiniLM-L6-v2", device="cuda")
```

```python
import torch
```

```python
from bertopic import BERTopic
```

```python
topic_model = BERTopic(
    embedding_model = 'all-MiniLM-L6-v2',
    min_topic_size=10
)
```

```python
topics, probs = topic_model.fit_transform(df['person1_venture_description'].tolist()) #docs is a list.
```

```python
df['topics']=topics
df['topic_probabilty']=probs
```

```python
df.to_csv('with_topic_modelling.csv')
```

```python
def get_top_3_topics(model, topic_id):
    words = model.get_topic(topic_id)
    if words is None:
        return ""
    return ", ".join([word for word, _ in words[:3]])
```

```python
cluster_to_name = {
    topic_id: get_top_3_topics(topic_model, topic_id)
    for topic_id in df["topics"].unique()
}
```

```python
df["top_topics"] = df["topics"].map(cluster_to_name)
```
